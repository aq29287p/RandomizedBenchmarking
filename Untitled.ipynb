```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
with open('fitting_data_K=100000.pkl', "rb") as f1:  
    K_10000_Prob_00 = pickle.load(f1)
    K_10000_m = pickle.load(f1)
f1.close()
def func(m, a,p, b):
    return a * p**m +b
popt, pcov = curve_fit(func, K_10000_m, K_10000_Prob_00)
y2 = [func(i, popt[0],popt[1],popt[2]) for i in K_10000_m]
plt.plot(K_10000_m,y2,'r--')
plt.plot(K_10000_m, K_10000_Prob_00, label='RB')
#plt.legend(loc='upper right')
#plt.xlabel('Sequence Length')
#plt.ylabel('Probability |00>')
#plt.savefig('D:/K=100000.jpg')
```
